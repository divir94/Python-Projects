```python
import datetime

class Bond:
    def __init__(self, payment, maturity, freq, nextCoupon, id):
        self.payment = payment
        self.maturity = datetime.datetime.strptime(maturity,"%Y%m%d").date()
        self.freq = freq
        self.nextCoupon = datetime.datetime.strptime(nextCoupon,"%Y%m%d").date()
        self.id = id
        
    #0 nextCoupon
    #1 1 coupon after that
    #-1 1 coupon in the past
    def couponDate(self, n):
        num_days = (12.0 / self.freq) * n * 30
        dt = datetime.timedelta(days=num_days)
        return self.nextCoupon + dt
    
    
    #paymentDate
    ##
    ### <--- today
    ####
    #####
    #paymentDate [#####] <---nextCouponDate(0)
    ## <--- accruedInterest
    ###
    #### 
    #####
    #paymentDate <--- nextCouponDate(1)
    ##
    ###
    ####     <---- TransactionDate ? accruedInterest
    #####
    #paymentDate <----- TransactioNDate ? accuredInterest <--nextCouponDate(2)
    ##
    ###
    ####
    #####
    #paymentDate + principal <--- maturity
    
    
    def partialAccruedInterest(self,transactionDate):
        interest = 0
        tempDate = self.nextCoupon
        period = 1
        while transactionDate < tempDate:
            tempDate = self.couponDate(period)
            interest += self.payment
            period += 1
        return interest
        
def bonds_from_arr(arr):
    bonds = [Bond(p, m, f, n, i) for (p, m, f, n, i) in arr]
    return bonds


def main():
    arr0 = ['x','r','d','e','f','b','r','m']
    
    # create a sorted array
    sortedArr0 = sorted(arr0, key = lambda x: ord(x))    
    # print sortedArr0

    # create a sorted array only with letters between 'd' and 'q'
    filteredSortedArr0 = [x for x in sortedArr0 if ord('d') <= ord(x) <= ord('q')]
    # print filteredSortedArr0

    arr1 = [(1.25,'20220610',2,'20160105','bond0'),
            (1.00,'20240610',2,'20160105','bond1'),
            (2.50,'20181210',4,'20160105','bond2'),
            (2.75,'20201210',2,'20160105','bond3'),
            (2.00,'20211210',4,'20160105','bond4'),
            (1.00,'20260610',2,'20160105','bond5'),
            (1.00,'20270610',4,'20160105','bond6'),
            (1.00,'20280610',2,'20160105','bond7'),
            (1.00,'20290610',2,'20160105','bond8')]

    bond0 = Bond(id='seansBond',payment=1.25,maturity='20220610',freq=2,nextCoupon='20160105')
    
    # create an array from the above that only includes values between 1.5 and 2.5
    bondArray = bonds_from_arr(arr1)
    # print bondArray
    
    print bond0.couponDate(2)

    # create a class from arr1 that represents a Bond
    # tuple1: coupon payment
    # tuple2: maturity date as a str
    # tuple3: Coupon Frequency per year
    # tuple4: Next coupon date
    # tuple5: unique bond identifier

    # Create a function for the class that calculates nextCoupon date and previousCouponDate 
    # n coupons into the future in either direction

    # Create a function that calculates the accrued interest at any date. When buying/selling a bond the accumulated accrued interest in this period is also settled with the transaction So if you are 10 days into a 90 day period you will receive 1/9 of the coupon

    # Create a class that holds a portfolio of the bond class above

    # Create the above classes in C++


if __name__=="__main__":
    main()
```
