```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv("evp.csv", index_col=[0])
df.head()
```

```python
@np.vectorize
def p2f(x):
    return float(x.strip('%'))/100

def calc_deviation(arr, mar, direction):
    deviation = arr[arr > mar] if direction == 'upside' else arr[arr < mar] 
    return np.sqrt(np.abs(deviation.sum()) / deviation.size)

for col_name in df.columns:
    col = p2f(df[col_name])
    print ("%s\n" % col_name) + "-"*10
    print "Upside: %.2f%%" % (100 * calc_deviation(col, col.mean(), 'upside'))
    print "Downside: %.2f%%\n" % (100 * calc_deviation(col, col.mean(), 'downside'))
```

```python
import pandas.io.data as web
```

```python
for ticker in ['^GSPC', 'ANF', 'SPY']:
    print "ticker: %s" % ticker
    data = web.DataReader(ticker, data_source='yahoo')
    start_date, end_date = '2012-12-31', '2014-05-23'
    data['Adj Close'][start_date : end_date].plot()
    data['Open'][start_date : end_date].plot()
    start, end = data['Adj Close'][start_date], data['Adj Close'][end_date]
    print "return: %.2f%%\n" % (100 * (end - start) / start)
    plt.show()
```
